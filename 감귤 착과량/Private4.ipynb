```python
import pandas as pd
import numpy as np
import os

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv(r'C:\Users\dlwks\OneDrive\바탕 화면\VSCode\DACON_CODE_REVIEW\감귤 착과량\train.csv')
test = pd.read_csv(r'C:\Users\dlwks\OneDrive\바탕 화면\VSCode\DACON_CODE_REVIEW\감귤 착과량\test.csv')
```

```python
train
```

```python
y_var = ['착과량(int)']
id_var = list(train.columns[2 : 6])
season = list(train.columns[6 : 95])
yuprokso = list(train.columns[95 :])
```

```python
train_melt = train.melt(id_vars = ['ID'] + y_var + id_var, value_vars = season + yuprokso)
test_melt = test.melt(id_vars = ['ID'] + id_var, value_vars = season + yuprokso)

train_melt['date'] = train_melt['variable'].apply(lambda x: x.split(' ')[0])
train_melt['type'] = train_melt['variable'].apply(lambda x: x.split(' ')[1])

test_melt['date'] = test_melt['variable'].apply(lambda x: x.split(' ')[0])
test_melt['type'] = test_melt['variable'].apply(lambda x: x.split(' ')[1])
```

```python
train_melt
```

```python
train_melt_preproc = train_melt[['ID'] + y_var + id_var + ['date']].drop_duplicates().reset_index(drop = True).merge(
    train_melt[train_melt['type'] == '새순'][['ID', 'date', 'value']].rename(columns = {'value' : '새순'}), how = 'left'
).merge(
    train_melt[(train_melt['type'] == '새순') & (train_melt['date'] == '2022-09-01')][['ID', 'value']].rename(columns = {'value' : '새순0'}), how = 'left'    
).merge(
    train_melt[(train_melt['type'] == '새순') & (train_melt['date'] == '2022-10-15')][['ID', 'value']].rename(columns = {'value' : '새순1'}), how = 'left'
).merge(
    train_melt[train_melt['type'] == '엽록소'][['ID', 'date', 'value']].rename(columns = {'value' : '엽록소'}), how = 'left'
)

train_melt_preproc['새순diff0'] = train_melt_preproc['새순'] - train_melt_preproc['새순0']
train_melt_preproc['새순diff1'] = train_melt_preproc['새순'] - train_melt_preproc['새순1']
train_melt_preproc['엽록소_새순'] = train_melt_preproc['엽록소'] / (train_melt_preproc['새순'] + 0.1) # 분모 0 방지
```

```python
test_melt_preproc = test_melt[['ID'] + id_var + ['date']].drop_duplicates().reset_index(drop = True).merge(
    test_melt[test_melt['type'] == '새순'][['ID', 'date', 'value']].rename(columns = {'value' : '새순'}), how = 'left'
).merge(
    test_melt[(test_melt['type'] == '새순') & (test_melt['date'] == '2022-09-01')][['ID', 'value']].rename(columns = {'value' : '새순0'}), how = 'left'
).merge(
    test_melt[(test_melt['type'] == '새순') & (test_melt['date'] == '2022-10-15')][['ID', 'value']].rename(columns = {'value' : '새순1'}), how = 'left'
).merge(
    test_melt[test_melt['type'] == '엽록소'][['ID', 'date', 'value']].rename(columns = {'value' : '엽록소'}), how = 'left'
)

test_melt_preproc['새순diff0'] = test_melt_preproc['새순'] - test_melt_preproc['새순0']
test_melt_preproc['새순diff1'] = test_melt_preproc['새순'] - test_melt_preproc['새순1']
test_melt_preproc['엽록소_새순'] = test_melt_preproc['엽록소'] / (test_melt_preproc['새순'] + 0.1) # 분모 0 방지
```

```python
display(train_melt_preproc)
display(test_melt_preproc)
```

```python
train_df = train_melt_preproc.pivot(index = ['ID'] + y_var + id_var, columns = ['date'], values = ['새순diff0', '새순diff0', '새순', '엽록소_새순']).reset_index()
test_df = test_melt_preproc.pivot(index = ['ID'] + id_var, columns = ['date'], values = ['새순diff0', '새순diff0', '새순', '엽록소_새순']).reset_index()
```

```python
display(train_df)
display(test_df)
```

```python
vars = []
for x in train_df.columns:
    if x[1] == '':
        vars.append(x[0])
    else:
        vars.append(x[0] + '' + x[1])

train_df.columns = vars
test_df.columns = ['ID'] + vars[2:]
```

```python
train_x, valid_x, train_y, valid_y = train_test_split(train_df.iloc[:, 2:], train_df[y_var], test_size = 0.4, random_state = 1740)
```

```python
rf = RandomForestRegressor(max_depth = 5, random_state = 1740, n_estimators = 200)
rf.fit(pd.concat([train_x, valid_x], axis = 0).iloc[:, 4:], pd.concat([train_y, valid_y], axis = 0).to_numpy().reshape(-1))
```

```python
pd.DataFrame({
    'ID' : test_df['ID'],
    '착과량(int)' : np.array(np.round(rf.predict(test_df.iloc[:, 5:]), 0), dtype = 'int')
}).to_csv('submit.csv', index = False)
```
